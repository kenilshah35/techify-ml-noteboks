```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn import preprocessing
```

```python
x = np.linspace(0, 2, 100)
fig,ax = plt.subplots()
plt.figure(figsize=(10,5))
ax.plot(x, x, label='linear')  
ax.plot(x, x**2, label='quadratic')  
ax.plot(x, x**3, label='cubic')  
ax.set_xlabel('x label')  
ax.set_ylabel('y label')  
ax.legend()
plt.show()
```

```python
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,5))
ax1.plot(t,data2,color='blue',label='sine')
ax3 = ax1.twinx()
ax3.plot(t,data1,color='red',label='exp')
ax1.legend()
ax3.legend()

ax2.plot(t,data2,label='sine')
ax2.set_aspect(1)
ax2.legend()
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,5))
np.random.seed(19680801)
t = np.arange(200)
x = 2**np.cumsum(np.random.randn(200))
ax.plot(t,x)
ax.set_yscale('log')
ax.set_xlim([20, 180])
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(20,7))
fruits = ['apple', 'orange', 'banana', 'kiwi']
counts = [40,30,100,20]
color = ['red', 'orange','yellow','green']

ax.bar(fruits, counts, color=color)
ax.set_ylabel('fruit supply')
ax.set_xlabel('fruit count')
ax.set_title('Fruits')
plt.show()
```

```python
df = pd.DataFrame(planets['mass'], columns = ['mass'])
df['distance'] = planets['distance']
stdscalecols = ['mass', 'distance']
df_norm = df[stdscalecols]
df_val = df_norm.values
stdscaler = preprocessing.MinMaxScaler()
df_scaled = stdscaler.fit_transform(df_val)
normalized_df = pd.DataFrame(df_scaled, columns = stdscalecols)
planets['mass'] = normalized_df['mass']
planets['distance'] = normalized_df['distance']

sns.pairplot(planets, diag_kind = 'kde')
```
